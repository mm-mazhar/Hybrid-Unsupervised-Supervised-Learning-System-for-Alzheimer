#### Load Data

In this part, created in file `1_eda.ipynb` is loaded:

```python
# Create a path object
# dataset_dir = cfgs["DATASET_DIR"]
dataset_dir = "..//dataset//modified"
dataset_path = Path(dataset_dir)
print(f"Dataset: {dataset_path}")

# # Find all text files inside a directory
# files = list(dataset_path.glob("*.csv"))

# Combining multiple paths
version = "01"
path_train_cleaned = dataset_path / f"train_v{version}.csv"
print(f"Train Data File Path --> {path_train_cleaned}")
```

```python
# Import the data as dataframe format
df = pd.read_csv(path_train_cleaned, encoding = 'utf8')
```

```python
# analyzing the dataset
report = sv.analyze([df, 'Train'], target_feat='composite_score')
```

```python
# show the report in a form of an HTML file
# VIS_DIR = cfgs["VISUALIZATION_DIR"]
VIS_DIR = "../visualization"
path_vis = Path(VIS_DIR)
print(path_vis)
path_visReport = f"{path_vis}/alzheimer_eda_v{version}.html"
print(f"EDA Report to Save: {path_visReport}")
print("\n")
report.show_html(path_visReport)
```
